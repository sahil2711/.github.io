Data Cleaning:
--------------

The NBA 2k20 game dataset originally contained multiple variations of individual players with varying attributes, each instance representing that player's abilities during a different season, with a different team, or merely for a special event (with their ratings usually boosted in order to encourage in-game purchases).

Choosing any of them at random may or may not be a true representation of a player's current ability. To eliminate this inconsistency, we opted to take the mean of every player's attributes across all variations. The logic behind this decision came from the observation that, for the most part, players' ratings would not vary all that wildly from variation to variation. Taking a median would give us an average, but it may not be entirely representative of a player's ability, since the same overall can be achieved with infinite permutations of the player's individual statistical ratings, and 2K has been known to manipulate these ratings in ways that reflect this. The mean as a measure of central tendency is known to be sensitive to outliers, however, the actual data used representing player ratings across their own iterations does not vary enough for this to make much of a difference.

Joining this newly trimmed data table with a table of 2019 NBA player statistics gave us a wide DataFrame from which we could isolate specific real-world and in-game statistics and run a battery of tests on them.

```python
#myteam statistics for each player represented in NBA2k is converted to a pandas dataframe via the read_csv method
myteam_players_dataframe = pd.read_csv('players.csv')


parameters_list = list(myteam_players_dataframe.columns)


#remove parameters not directly related to ovr calculation 
parameters_list.remove("id")
parameters_list.remove("theme")
parameters_list.remove("team")
parameters_list.remove("tier")
parameters_list.remove("position")
parameters_list.remove("secondary_position")
parameters_list.remove("play1")
parameters_list.remove("play2")
parameters_list.remove("play3")
parameters_list.remove("play4")

#remove parameters that represent parts of NBA2k's mircotransaction economy instead of meaningful statistics
parameters_list.remove("bronze_badges")
parameters_list.remove("silver_badges")
parameters_list.remove("gold_badges")
parameters_list.remove("hof_badges")
parameters_list = parameters_list[:parameters_list.index("hard_foul_t")+1]


#remove null/infinite values from dataframe
myteam_players_dataframe = myteam_players_dataframe[parameters_list]
myteam_players_dataframe = myteam_players_dataframe.dropna()

#remove name of player and their overall score from the parameters 
parameters_list.remove("name")
parameters_list.remove("overall")
```

```python
#reads csv of NBA 2018-2019 season stats and converts it to a dataframe
nba_players_dataframe = pd.read_csv('IRLNBA_ALL1819.csv')
nba_players_dataframe.head(10)

nba_with_2K_data = pd.merge(myteam_players_dataframe,nba_players_dataframe, on='name')

#groups NBA2K and real-life NBA stats together by name, and takes the mean of the 2K stats for players with more than one observation
nba_with_2K_data = nba_with_2K_data.groupby('name').mean().reset_index()

nba_with_2K_data.head(5)
```

Data Testing:
-------------

To make sense of the newly built Dataframe, we decided to use multivariate-linear regression to train and test our data across 8 key parameters: Three-point Field Goal Completion Percentage, Two-point Field Goal Completion Percentage, Points per game, True Shooting Percentage, Rebounds per game, Assists per game, Steals per game, Blocks per game, and Free Throw percentage. Since the variables are continuous and independent in nature, linear regression is the best option to find a relationship between predicted and actual values. Each key parameter has been tested individually using the categorization described above.

We measured the accuracy of the multivariate linear regression model for each key parameter via 10 fold cross-validation, calculating the mode's coefficient of determination, and calculating the model's correlation coefficient.

K-Fold cross-validation is when we randomly divide into k groups or folds of equal size, with the first fold treated as a validation set and trained on the remaining k-1 folds. The mean squared error is computed on the validation and this process to repeated k times. Each time, a different set of observations is treated as the validation set. To obtain the accuracy of our multivariate linear regression model (mean squared error), we take the mean of the score outputted by our cross-validation method.

The coefficient of determination is a statistic that gives the proportion (%) of the variation in y-values that can be explained by variation in the x-values. For example, if r2=0.78that means 78% of the variation in y-values can be explained in x-values, and 22% of the variation in y-values can be unexplained by variation in the x-values. This statistic can be obtained by calling the sklearn 'score' function on the multivariate linear regression model with the training data as arguments.

The correlation coefficient is the square root of the coefficient of determination. The correlation coefficient is used to measure how strong the relationship between the independent and dependent variables is.  The correlation coefficient is a value between 1 and -1, where 1 indicates a strong positive relationship, -1 indicates a strong negative relationship, and 0 indicates no relationship at all.

Data Visualization:
-------------------

For each real-life basketball stat predicted, we start by creating two scatter plots. First, we plot the estimated values from the linear regression model for that specific stat against the actual values in a scatter plot. Then, we plot the residuals to determine how much the estimated values vary about the mean. From these two plots, we can get an idea of the strength and accuracy of our model, respectively.

```python
#Multivariate Linear Regression Model for 3P%

#training data
X = nba_with_2K_data[['potential','hustle','intangibles','shot_3pt', 'spot_up_shot_3pt_t', 'shot_3pt_t', 'spot_up_shot_3pt_t', 'off_screen_shot_3pt_t', 'drive_pull_up_3pt_t', 'stepback_jumper_3pt_t','transition_pull_up_3pt_t', 'drive_pull_up_3pt_t']]
Y = nba_with_2K_data['3PointPercent']

#create multivariate linear regression model
lin_reg = LinearRegression().fit(X, Y)
#calculates the 10-cross fold variation of the model
scores = cross_val_score(lin_reg, X, Y, cv=10)
#outputs accuracy of the model according to 10 cross fold variation score
print("Model Accuracy: ", scores.mean())
#print("Actual v. Predicted Results Paired T-Test:", stats.ttest_rel(Y,lin_reg.predict(X)))
#outputs the coefficient of determination of the model
print("Coefficient of Determination:", lin_reg.score(X,Y))
print("Correlation Coefficient:", math.sqrt(lin_reg.score(X,Y)))

#list of predicted values
x_pred = lin_reg.predict(X)
#slope of the linear regresiion model
slope = lin_reg.coef_[0]
#y-intercept of the model
intercept = lin_reg.intercept_
#point-slope form, used to graph line of best fit
y_line = intercept + x_pred*slope

predicted_vs_actual = {'Actual':Y,'Predicted':x_pred, 'Residual':Y-x_pred}
fig, ax = plt.subplots(2, figsize=(8,16))
ax[0].set_title('3-Point shot Percentage Multivariate Linear Regression')
ax[0].set_xlabel('Predicted values')
ax[0].set_ylabel('Actual values')
ax[0].scatter(x_pred, Y)
ax[1].set_title('Residuals Graph')
ax[1].set_xlabel('Predicted values')
ax[1].set_ylabel('Residual values')
ax[1].scatter(x_pred, Y-x_pred)
ax[1].axhline(0)


df = pd.DataFrame(predicted_vs_actual)
df.head(5)
```

Analysis:
-----------


Based on the Actual values vs Predicted values scatterplots and the Residual values scatterplots, we can see that some real-life NBA stats do have strong linear relationships with their representative statistics in NBA2K. The point distribution and shape for the Actual values vs Predicted values scatterplots and the Residual values scatterplots for Points Per Game (PPG), Assists Per Game (AssistsPG), and Rebounds Per Game (RPG) indicate that there is a strong linear relationship between these real-life NBA player statistics and specific player statistics in NBA2K. 

The correlation coefficients for Points Per Game (PPG), Assists Per Game (AssistsPG), and Rebounds Per Game (RPG) are 0.8916514848782566, 0.7830341650216279, and 0.7613072042533633 respectively. A correlation coefficient greater than or equal to 0.7 typical indicates a strong linear relationship between the independent and dependent variables. Thus, we can conclude that Points Per Game (PPG), Assists Per Game (AssistsPG), and Rebounds Per Game (RPG) (dependent variables) have a strong linear relationship with their representative NBA2K statistics (independent variables). 

Based on the shape of their Actual values vs Predicted values scatterplots and the Residual values scatterplot for 2P%, Steals, and BPG appears to have moderate linear relationships with NBA2K statistics. 

The correlation coefficients for the 2P%, Steals, and BPG models are 0.5801701635949479, 0.6419007818561976, and 0.646030987453557 respectively. The values of the aforementioned correlation coefficient indicate that 2P%, Steals, and BPG (dependent variables) do have moderate linear relationships with NBA2K statistics (independent variables). 

However, the remainder of the real-life NBA2K stats appears to have a weak or no linear relationship with their representative NBA2K statistics. The point distribution and shape for the Actual values vs Predicted values scatterplots and the Residual values scatterplots for 3P% and Free Throw % indicates that there is a weak linear relationship or no relationship between these real-life statistics and their representative statistics in NBA2K.

The correlation coefficients for the 3P% and Free Throw % models are 0.4937702103631969 and 0.3055693414986993 respectively. The values of these correlation coefficients indicate that 3P% and Free Throw% (dependent variables) have a weak linear relationship with NBA2K statistics (independent variables). 

Conclusion:
-----------

The data analysis yields the conclusion that there is a relationship between real life NBA statistics and in-game statistics in NBA2K.  6 of the 8 key real-life NBA statistics that were analyzed -Points Per Game (PPG), Assists Per Game (AssistsPG), Rebounds Per Game (RPG), 2P%, Steals, and BPG (Blocks Per Game- had strong or moderate correlations with the in-game player statistics in NBA2K.  However, two real-life NBA statistics - 3P% and Free Throw % - had weak correlations with the in-game statistics in NBA2K.

Overall, we can conclude that the in-game player statistics in NBA 2K do have some basis in real-life NBA player statistics. However, further analysis will need to be done before it can be determined how the developers translate real-life NBA player statistics into NBA2K in-game player statistics. It remains unclear how certain, more abstract in-game statistics such as potential, intangibles, and hustle are derived from real-life NBA statistics. It also remains unclear how much in-game statistics are directly altered by developers for balance/gameplay purposes, which may explain why some real-life statistics have only a moderate or weak correlation to in-game statistics.
